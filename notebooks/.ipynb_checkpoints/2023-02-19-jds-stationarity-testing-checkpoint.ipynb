```python
arima_24_0_0_unfitted = ARIMA(train_data, order=(24, 0, 0))
start = time.perf_counter()
arima_24_0_0 = arima_24_0_0_unfitted.fit()
print(f'Took {time.perf_counter() - start}s to train arima 24')
```

```python
with open('../data/04_models/arima_24_0_0.obj', 'wb') as file:
    start = time.perf_counter()
    pickle.dump(arima_24_0_0, file)
print(f'Took {time.perf_counter() - start}s to pickle arima(24, 0, 0)')
```

```python
with open('../data/04_models/arima_24_0_0.obj', 'rb') as file:
    arima_24_0_0 = pickle.load(file)
```

```python
date_range = np.arange(datetime(2021,1,1,0), datetime(2021,12,31,23), timedelta(hours=1))

y = ercot.loc[datetime(2022, 1, 1, 0):datetime(2022,12,31,23)].values

y_hat = np.asarray([arima_24_0_0.predict(date).values[0] for date in date_range])
```

```python
plt.plot(date_range, y, color='blue')
plt.plot(date_range, y_hat, color='green')
plt.savefig('arma_24_0_0_2022_predictions.png')
```

```python
residuals = y - y_hat
plt.plot(date_range, residuals)
plt.title('Residual Plot')
plt.savefig('arma_24_0_0_residual.png')
```
